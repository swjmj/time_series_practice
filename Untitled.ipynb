```python
#Método 2...moving average; en este método se toma el promedio de conteo de solamente los últimos periodos 
#rolling es un método que sólo se puede usar con series de tiempo (teniendo un índice Timestamp)
#este método te da una ventana de observaciones para hacer calculos, esto es toma de x en x observaciones o datos


y_hat_avg = valid.copy()
y_hat_avg['moving_avg_forecast'] = Train.Count.rolling(10).mean().iloc[-1]
rms = sqrt(MSE(valid.Count, y_hat_avg.moving_avg_forecast))

plt.figure(figsize=(15, 5))
plt.plot(Train.Count, label='Entrenamiento')
plt.plot(valid.Count, label='Vaidacion')
plt.plot(y_hat_avg.moving_avg_forecast, label='Usando las últimas 10 observaciones: RMSE='+str(rms))
plt.legend(loc='best')

y_hat_avg = valid.copy()
y_hat_avg['moving_avg_forecast'] = Train.Count.rolling(20).mean().iloc[-1]
rms = sqrt(MSE(valid.Count, y_hat_avg.moving_avg_forecast))

plt.figure(figsize=(15, 5))
plt.plot(Train.Count, label='Entrenamiento')
plt.plot(valid.Count, label='Vaidacion')
plt.plot(y_hat_avg.moving_avg_forecast, label='Usando las últimas 20 observaciones: RMSE='+str(rms))
plt.legend(loc='best')

y_hat_avg = valid.copy()
y_hat_avg['moving_avg_forecast'] = Train.Count.rolling(50).mean().iloc[-1]
rms = sqrt(MSE(valid.Count, y_hat_avg.moving_avg_forecast))

plt.figure(figsize=(15, 5))
plt.plot(Train.Count, label='Entrenamiento')
plt.plot(valid.Count, label='Vaidacion')
plt.plot(y_hat_avg.moving_avg_forecast, label='Usando las últimas 50 observaciones: RMSE='+str(rms))
plt.legend(loc='best')
plt.show()
```

```python
#Simple exponential smoothing: Aquí es la misma idea de promedios, pero se asignan pesos más grandes a observaciones
#más recientes y pesos más pequeños a observaciones de hace tiempo
#Los pesos decrecen exponencialmente mientras más lejanas(en el sentido temporal) son las observaciones; los pesos 
#más pequeños son asignados a las observaciones más viejas

from statsmodels.tsa.api import ExponentialSmoothing, SimpleExpSmoothing, Holt

y_hat_avg = valid.copy()
fit2 = SimpleExpSmoothing(np.asarray(Train.Count)).fit(smoothing_level=0.6, optimized=False)
y_hat_avg['SES'] = fit2.forecast(len(valid))
rms = sqrt(MSE(valid.Count, y_hat_avg.SES))

plt.figure(figsize = (16, 8))
plt.plot(Train.Count, label='Entrenamiento')
plt.plot(valid.Count, label='Validación')
plt.plot(y_hat_avg.SES, label='SES '+str(rms))
plt.legend(loc='best')
plt.show()
```

```python
"""
    Introducción al modelo ARIMA
    Para poder utilizar este modelo, es necesario que la serie de tiempo sea estacionaria. Si no es estacionaria
    se hace estacionaria
    
    Para probar que la seria es estacionaria, utilizaremos la prueba de Dickey Fuller
    
    Esta prueba arroja como resultado "Test Statistic y algunos Valores criticos (Critical Values) para diferentes
    niveles de confianza. Si "Test Statistic" es menor que "Critical value" podemos rechazar la hipótesis nula 
    (la serie no es estacionaria) y decir que la seria es estacionaría
"""


from statsmodels.tsa.stattools import adfuller

def test_stationarity(timeseries):
    #determing rolling statistics
    rolmean = timeseries.rolling(window=24).mean() #24 por las horas en un dia...es la ventana para promedia
    rolstd = timeseries.rolling(window=24).std()
    #graficando 
    orig = plt.plot(timeseries, color='blue', label='Original')
    mean = plt.plot(rolmean, color='red', label='Rolling Mean')
    std = plt.plot(rolstd, color='black', label='Rolling Std')
    plt.legend(loc='best')
    
    #haciendo la prueba de Dickey Fuller
    print('Resultados de la prueba de Dickey Fuller:')
    dftest = adfuller(timeseries, autolag='AIC')
    dfoutput = pd.Series(dftest[0: 4], index=['Test Statistic', 'p-value', '#Lags usados','Numero de observaciones usados'])
    for key, value in dftest[4].items():
        dfoutput['Valor critico (%s)' %key] = value
    print(dfoutput)
    
from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 20, 10
test_stationarity(train_original['Count'])
    
```

Como se puede ver 'Test Statistic' es menor que el valor crítico de 1%, pero se puede ver que la serie de tiempo tiene una
tendencia.

Una tendencia existe cuando la serie de tiempo muestra un crecimiento o decrecimiento a largo plazo. Este crecimiento no es precisamente lineal.

Cuando vemos una tendencia a  crecer en los datos podemos aplicar alguna tranformación. Aquí aplicaremos una tranformación que penaliza valores grandes mucho más que valores pequeños, por ejemplo aplicando un logaritmo.

Otra cosa que se puede hacer para eliminar la tenedencia es remover el "Rolling average" de la serie de tiempo.

Lo siguiente es descomponer la serie en tendencia y estacionalidad, con esto obtendremos el residuo que es la variación aleatoria de la serie de tiempo.

Estacionalidad:

Con estacionalidad nos referimos a las fluctuaciones periodicas. Un patron estacional existe cuando la serie está influenciada por factores estacionales (ejem. las estaciones del año, el mes o dia de la semana)

La estacionalidad es siempre un periodo conocido y fijo  

Usaremos descomposición estacional para descomponer la serie de tiempo en tendencia, estacionalidad y residuo.

```python
from statsmodels.tsa.seasonal import seasonal_decompose

decomposition = seasonal_decompose(pd.DataFrame(Train_log).Count.values, freq=24)

trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid

plt.subplot(411)
plt.plot(train_log, label='Original log')
plt.legend(loc='best')
plt.subplot(412)
plt.plot(trend, label='Trend')
plt.legend(loc='best')
plt.subplot(413)
plt.plot(seasonal, label='Seasonality')
plt.legend(loc='best')
plt.subplot(414)
plt.plot(residual, label='Residual')
plt.legend(loc='best')
plt.tight_layout()
plt.show()
```

```python
#Ahora comprobemos la estacionalidad de los residuos

train_log_decompose = pd.DataFrame(residual)
train_log_decompose['date'] = Train_log.index
train_log_decompose.set_index('date', inplace=True)
train_log_decompose.dropna(inplace=True)
train_log_decompose.head(50)
test_stationarity(train_log_decompose[0])
```

Como es fácil ver, "Test Statistic" es un valor mucho menor al Valor Crítico de 1%, podemos concluir que la seríe de tiempo es estacionaría. 
Es hora de hacer forecasting usando ARIMA

Lo primero que tenemos que hacer es ajustar el modelo de ARIMA a nuestra seríe de tiempo para lo cual tenemos que encontrar los valores optimos para los parámetros p, d, q.

Para encontrar estos parámetros optimos usaremos las graficas de ACF (Autocorrelation function) y PACF (Partial Autocorrelation Function).

ACF es una médida de la correlación de la serie de tiempo con una seríe retrazada de si misma.

PACF mide la correlación entre la serie de tiempo con una versión retrazada de si misma pero después de eliminadas las comparaciones intermedias (intervening comparisons)

```python
from statsmodels.tsa.stattools import acf, pacf
lag_acf = acf(train_log_diff.dropna(), nlags = 25)
lag_pacf = pacf(train_log_diff.dropna(), nlags=25, method='ols')
```

```python
plt.plot(lag_acf)
plt.axhline(y=0, linestyle='--', color='gray')
plt.axhline(y=-1.96/np.sqrt(len(train_log_diff.dropna())), linestyle='--', color='gray')
plt.axhline(y=1.96/np.sqrt(len(train_log_diff.dropna())), linestyle='--', color='gray')
plt.title('Funcion de autocorrelación')
plt.show()
plt.plot(lag_pacf)
plt.axhline(y=0, linestyle='--', color='gray')
plt.axhline(y=-1.96/np.sqrt(len(train_log_diff.dropna())), linestyle='--', color='gray')
plt.axhline(y=1.96/np.sqrt(len(train_log_diff.dropna())), linestyle='--', color='gray')
plt.title('Funcion de autocorrelación parcial')
plt.show()
```

Los valores de p y q se obtienen fijandonos en el primer cruce de cada curva con el intervalo de confianza, en este caso para la curva ACF q=1 y para la curva PACF es p=1.

Ahora que tenemos los valores de p y q podemos hacer el modelo de ARIMA, primero haremos el modelo AR y después el MA finalmente los combinaremos.

```python
#Modelo AR, en este modelo se asume que la variable de salida (output) depende linealmente de sus propios valores previos.

from statsmodels.tsa.arima_model import ARIMA

model = ARIMA(Train_log, order=(2, 1, 0)) #aqui el valor de q es cero ya que sólo queremos el modelo AR
results_AR = model.fit()

plt.plot(train_log_diff.dropna(), label='original')
plt.plot(results_AR.fittedvalues, color='red', label='predicción')
plt.legend(loc='best')
plt.show()

pd.DataFrame(results_AR.resid).plot( kind='kde', label='residuos')
plt.show()

#results_AR.plot_diagnostics(figsize=(16, 8))
plt.show()
print(pd.DataFrame(results_AR.resid).describe())
```

```python
#Graficamos la curva de validación para el modelo AR
#Tenemos que hacer la transformación inversa el modelo
#Primero hagamos una predicción y guardemosla en una variable aparte

AR_predict = results_AR.predict(start='2014-06-25', end='2014-09-25') #se realiza la predición
AR_predict = AR_predict.cumsum().shift().fillna(0) #cumsum() hace la suma acumulada...es como agregar rolling average shift mueve 1 índice AR_predict
AR_predict1 = pd.Series(np.ones(valid.shape[0])*np.log(valid.Count)[0], index= valid.index) #se hace una serie de ceros y se multiplica por el primer valor del log del set de validación
AR_predict1 = AR_predict1.add(AR_predict, fill_value=0) #se agrega el vector anterior a AR_predict para "levantarlo"
AR_predict = np.exp(AR_predict1) #se saca la exponencial

plt.plot(train.Count, label = 'Validación')
plt.plot(valid.Count, label = 'Validación')
plt.plot(AR_predict, color= 'red', label = 'Prediccion')
plt.legend(loc='best')
plt.title('Error cuadrático medio: %.4f'%(np.sqrt(np.dot(AR_predict, valid.Count))/valid.shape[0]))
plt.show()
```

```python
#para no repetir codigo una funcion que me regresa a la escala inicial
def return_scale(predicted_values, valid):
    predicted_values = predicted_values.cumsum().shift().fillna(0) #cumsum() hace la suma acumulada...es como agregar rolling average shift mueve 1 índice AR_predict
    predicted_values1 = pd.Series(np.ones(valid.shape[0])*np.log(valid.Count)[0], index= valid.index) #se hace una serie de ceros y se multiplica por el primer valor del log del set de validación
    predicted_values1 = predicted_values1.add(predicted_values, fill_value=0) #se agrega el vector anterior a AR_predict para "levantarlo"
    predicted_values = np.exp(predicted_values1) #se saca la exponencial

    plt.plot(valid.Count, label = 'Validación')
    plt.plot(predicted_values, color= 'red', label = 'Prediccion')
    plt.legend(loc='best')
    plt.title('Error cuadrático medio: %.4f'%(np.sqrt(np.dot(predicted_values, valid.Count))/valid.shape[0]))
    plt.show()
```
