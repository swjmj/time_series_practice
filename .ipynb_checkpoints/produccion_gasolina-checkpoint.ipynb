```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
from statsmodels.tsa.api import ExponentialSmoothing, SimpleExpSmoothing, Holt
from sklearn.metrics import mean_squared_error as MSE
from IPython.core.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))
```

```python
gas = pd.read_csv('resumen_gasolinas.csv')
```

```python
gas.columns
gas.dtypes
gas.shape
```

```python
gas.head()
```

```python
gas['year'] = pd.to_datetime(gas.year, format='%Y')
gas['produccion_copy'] = gas['produccion']
gas['produccion_org'] = gas['produccion']*10**9/34.78
gas['produccion'] = gas['produccion']/np.sum(gas['produccion'].values)
```

```python
gas.index = gas.year
gas.head()
```

```python
plt.figure(figsize = (28, 16))
plt.plot(gas.produccion_org, label='Producción x $10^{10}$ litros de Gasolina')
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.title('Serie de tiempo')
plt.xlabel('Tiempo (año)', fontsize=30)
plt.ylabel('x $10^{10}$ litros de Gasolina', fontsize=30)
plt.legend(loc='best', fontsize=30)
plt.title('Producción de gasolina', fontsize=50)
plt.show()
```

```python
#descomposición estacionaria de la serie para la importación de gasolina
import statsmodels.api as sm
decompose = sm.tsa.seasonal_decompose(gas.produccion)

trend = decompose.trend
seasonality = decompose.seasonal
residual = decompose.resid

plt.figure(figsize=(15, 10))
plt.subplot(411)
plt.plot(gas.produccion, label='original')
plt.legend(loc='best')
plt.subplot(412)
plt.plot(trend, label='trend')
plt.legend(loc='best')
plt.subplot(413)
plt.plot(seasonality, label='seasonality')
plt.legend(loc='best')
plt.subplot(414)
plt.plot(residual, label='residual')
plt.legend(loc='best')
plt.tight_layout()
plt.legend(loc='best')
plt.show()
```

```python
#haciendo la prueba de Dickey Fuller
print('Resultados de la prueba de Dickey Fuller:')
dftest = sm.tsa.stattools.adfuller(gas.produccion, autolag='AIC')
dfoutput = pd.Series(dftest[0: 4], index=['Test Statistic', 'p-value', '#Lags usados','Numero de observaciones usados'])
for key, value in dftest[4].items():
    dfoutput['Valor critico (%s)' %key] = value
print(dfoutput)
```

```python
#lo primero que voy a hacer es usar Holt

train = gas.ix['1991':'2008']
valid = gas.ix['2008':'2017']

plt.figure(figsize=(16, 8))
plt.plot(train.produccion)
plt.plot(valid.produccion)
plt.show()
```

```python
temp = []
l=100
for i in range(l):
    for j in range(l):
        fit1 = Holt(np.asarray(train.produccion)).fit(smoothing_level = i/l, smoothing_slope = j/l)
        valid['Holt_linear'] = fit1.forecast(len(valid))
        rms = np.sqrt(MSE(valid.produccion, valid.Holt_linear))
        temp.append([i/l, j/l, rms])
        
best = np.argmin(np.asarray(temp)[:, 2])
print(temp[best])
```

```python
fit1 = Holt(np.asarray(train.produccion)).fit(smoothing_level = 0.89, smoothing_slope = 0.99)
valid['Holt_linear'] = fit1.forecast(len(valid))
rms = np.sqrt(MSE(valid.produccion, valid.Holt_linear))
plt.figure(figsize=(28, 16))
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.plot(train.produccion, label='Entrenamiento')
plt.plot(valid.produccion, label='Validación')
plt.plot(valid.Holt_linear, label=r'Holt Linear, RMS= %.5f con $\alpha = 0.89$ y $\beta^*=0.99$'%rms)
plt.xlabel('Tiempo (años)', fontsize=30)
plt.legend(loc='best', fontsize=25)
plt.show()
```

```python
prediction_time = pd.date_range('2008', '2020', freq='Y')
prediction_years = pd.DataFrame(index=prediction_time)
prediction_years['year'] = prediction_time.year

prediction_years['prediction'] = fit1.forecast(len(valid)+2)
```

```python
plt.figure(figsize=(16, 8))
plt.plot(prediction_years.prediction, label='prediccion')
plt.plot(gas.produccion, label='original')
plt.legend(loc='best')
plt.show()
```

```python
#prediciendo usando toda la serie como entrenamiento

fit1 = Holt(np.asarray(gas.produccion)).fit(smoothing_level = 0.89, smoothing_slope = 0.99)
pred_range = pd.date_range('2018','2020', freq ='YS')
forecast = pd.DataFrame(index=pred_range)
forecast['pred'] = fit1.forecast(3)

plt.figure(figsize=(28, 16))
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.plot(gas.produccion*(10**9/34.78)*np.sum(gas['produccion_copy']), label='$10^{10}$ Litros de Gasolina')
plt.plot(forecast['pred']*(10**9/34.78)*np.sum(gas['produccion_copy']), label=r'Predicción años 2018=%.3f $\times 10^{10}$ , 2019=%.3f  $\times 10^{10}$ y 2020=%.3f  $\times 10^{10}$ usando el método de Holt Linear'
         %((forecast['pred']*(1/(34.78*10))*np.sum(gas['produccion_copy']))[0], (forecast['pred']*(1/(34.78*10))*np.sum(gas['produccion_copy']))[1], (forecast['pred']*(1/(34.78*10))*np.sum(gas['produccion_copy']))[2]))
plt.xlabel('Tiempo (año)', fontsize=30)
plt.ylabel('$10^{10}$ Litros de Gasolina', fontsize=30)
plt.legend(loc='best', fontsize=20)
#plt.title('Producción de gasolina - $10^{10}$ litros de gasolina - Prediciendo con toda la serie', fontsize=30)
plt.show()
```

```python
#haciendo la prueba de Dickey Fuller, se podrá aplicar ARIMA?
print('Resultados de la prueba de Dickey Fuller:')
tranf_series1 = ((gas.produccion)/gas.produccion.cumsum()).dropna()
tranf_series = (gas.produccion-gas.produccion.shift(1)).dropna()

tranf_series_rolling = (gas.produccion-gas.produccion.rolling(window=6).mean()).dropna()

dftest = sm.tsa.stattools.adfuller(tranf_series, autolag='AIC')
dfoutput = pd.Series(dftest[0: 4], index=['Test Statistic', 'p-value', '#Lags usados','Numero de observaciones usados'])
for key, value in dftest[4].items():
    dfoutput['Valor critico (%s)' %key] = value
print(dfoutput)

plt.figure(figsize=(16, 8))
plt.plot(tranf_series)
plt.show()
```

```python
#lo primero que voy a hacer es usar Holt

train = gas.ix['1991':'2012']
valid = gas.ix['2012':'2017']

plt.figure(figsize=(16, 8))
plt.plot(train.produccion)
plt.plot(valid.produccion)
plt.show()
```

```python
#Curvas ACF y PACF

from statsmodels.tsa.stattools import acf, pacf
lag_acf = acf(train.produccion, nlags = 10)
lag_pacf = pacf(train.produccion, nlags=10, method='ols')
```

```python
plt.figure(figsize=(16, 8))
plt.plot(lag_acf)
plt.axhline(y=0, linestyle='--', color='gray')
plt.axhline(y=-1.96/np.sqrt(len(train)), linestyle='--', color='gray')
plt.axhline(y=1.96/np.sqrt(len(train)), linestyle='--', color='gray')
plt.title('Funcion de autocorrelación')
plt.show()
plt.figure(figsize=(16, 8))

plt.plot(lag_pacf)
plt.axhline(y=0, linestyle='--', color='gray')
plt.axhline(y=-1.96/np.sqrt(len(train)), linestyle='--', color='gray')
plt.axhline(y=1.96/np.sqrt(len(train)), linestyle='--', color='gray')
plt.title('Funcion de autocorrelación parcial')
plt.show()
```

```python
from statsmodels.tsa.arima_model import ARIMA

model_MA = ARIMA(train.produccion, order=(0, 2, 1))
fitted_MA = model_MA.fit()
print(fitted_MA.summary())

fig, ax = plt.subplots(figsize=(16, 8))
train.produccion.plot(ax=ax)
valid.produccion.plot(ax=ax)
ax.vlines('2012', 0, 1000, linestyle='--', color='r', label='Inicio de la predicción')
#plt.plot(fitted_MA.fittedvalues, label='fitted')
#plt.plot(fitted_MA.resid.values, label='residuos')
ax = fitted_MA.plot_predict(start='2012', end='2017', dynamic=True, plot_insample=False, ax=ax)


#plt.legend(loc='best')
```

```python
model_fitted = pd.DataFrame({'predicted' : fitted_MA.predict(start='2012', end='2017', dynamic=True).values})
model_fitted.index = fitted_MA.predict(start='2012', end='2017', dynamic=True).index
model_fitted = model_fitted.cumsum().shift().fillna(0).add(valid.produccion.iloc[0])


rme =np.sqrt(MSE(valid.produccion, model_fitted))


plt.figure(figsize=(16, 8))
plt.plot(gas.produccion, label='importacion')
plt.plot(model_fitted, label='forecast')
plt.title('El error es: '+str(rme))
plt.legend(loc='best')
plt.show()
```
